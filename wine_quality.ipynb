```python
sb.pairplot(df_wine,hue='target',vars=['alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash', 'magnesium',
       'total_phenols', 'flavanoids', 'nonflavanoid_phenols',
       'proanthocyanins', 'color_intensity'])
```

```python
sb.countplot(df_wine['target'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df_wine.drop(['target'],axis=1)
y=df_wine['target']
```

```python
X1,X2,y1,y2=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X1,y1)
```

```python
pr=lr.predict(X2)
```

```python
pr
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y2,pr)
```

```python
from sklearn.cluster import KMeans
```

```python
elb=[]
for i in range (1,5):
    km=KMeans(n_clusters=i)
    km.fit(X)
    elb.append(km.inertia_)
```

```python
elb
```

```python
plt.plot(range(1,5),elb)
plt.scatter(range(1,5),elb)
```

```python
fkm=KMeans(n_clusters=2)
fkm.fit(X)
```

```python
fkm.cluster_centers_
```

```python
pr=fkm.predict(X)
```

```python
pr
```

```python
plt.scatter(X[pr==0,0],X[pr==0,0],label="First Group")
plt.scatter(X[pr==1,0],X[pr==1,0],label="Second Group")
plt.scatter(fkm.cluster_centers_[:,0],fkm.cluster_centers_[:,0],label="Cluster")
plt.legend()
```
